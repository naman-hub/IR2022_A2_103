```python
def ndcg(df, n):
    ans = 0
    for i in range(n):
        ans += (pow(2,df[0][i]) - 1)/np.log2(i+2)
    return ans
```

```python
import matplotlib.pyplot as plt
```

```python
relevant = 0
for e in arr_3:
    if e != 0:
        relevant+=1

c = 0
r = 0
precision = []
recall = []



for e in arr_3:
    c += 1 
    if e!=0:
        r+=1
    precision.append(r/c)
    recall.append(r/relevant)
plt.plot(recall,precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.show()
```
